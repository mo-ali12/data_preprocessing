```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../Jupyter/Datasets/train.csv')
df.head()
```

```python
df.describe(include = 'all')
```

```python
missing_val = df.isnull()
for c in missing_val.columns.values.tolist():
    print(c)
    print(missing_val[c].value_counts())
    print("")
```

```python
mean = df['Age'].astype('float').mean()
df['Age'].replace(np.nan, mean, inplace = True)
```

```python
del df['Name']
del df['Ticket']
del df['Fare']
del df['Cabin']
```

```python
df
```

```python
freq = df['Embarked'].value_counts().idxmax()
df.replace(np.nan,freq, inplace = True)
df
```

```python
def getNumber(str):
    if str == 'male':
        return 1
    else:
        return 2
df['Gender'] = df['Sex'].apply(getNumber)
df.head()
```

```python
#Drawing a pie chart for number of males and females aboard
import matplotlib.pyplot as plt
from matplotlib import style

males = (df['Sex'] == 'male').sum() 
#Summing up all the values of column gender with a 
#condition for male and similary for females
females = (df['Sex'] == 'female').sum()
print(males)
print(females)
p = [males, females]
plt.pie(p,    #giving array
       labels = ['Male', 'Female'], #Correspndingly giving labels
       colors = ['green', 'yellow'],   # Corresponding colors
       explode = (0.15, 0),    #How much the gap should me there between the pies
       startangle = 0)  #what start angle should be given
plt.axis('equal') 
plt.show()
```

```python
# More Precise Pie Chart
MaleS=df[df.Sex=='male'][df.Survived==1].shape[0]
print(MaleS)
MaleN=df[df.Sex=='male'][df.Survived==0].shape[0]
print(MaleN)
FemaleS=df[df.Sex=='female'][df.Survived==1].shape[0]
print(FemaleS)
FemaleN=df[df.Sex=='female'][df.Survived==0].shape[0]
print(FemaleN)
```

```python
chart=[MaleS,MaleN,FemaleS,FemaleN]
colors=['lightskyblue','yellowgreen','Yellow','Orange']
labels=["Survived Male","Not Survived Male","Survived Female","Not Survived Female"]
explode=[0,0.05,0,0.1]
plt.pie(chart,labels=labels,colors=colors,explode=explode,startangle=100,counterclock=False,autopct="%.2f%%")
plt.axis("equal")
plt.show()
```
